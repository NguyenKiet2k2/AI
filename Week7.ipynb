```python
import pandas as pd
url = 'mtcars.csv'
data= pd.read_csv(url)
import matplotlib.pyplot as plt # Thư viện dùng cho machine learning - mô hình tuyến tính
data.hist() #vẽ phân bố dữ liệu Histogram
plt.show()

"""
# mpg : so quảng đường đi được trên 1 gal nhiên liệu
wt: khoi luong cua xe
# + thì tỉ lệ thuận; - là tỉ lệ nghịch 
giá trị tương quang giữa nó với chính nó  =11

"""
correlation= data.corr()
print(correlation)
```

```python
import seaborn as sns
sns.heatmap(correlation, annot= True)
```

```python
data.plot(kind='scatter', x='wt', y='mpg',figsize=(9,9),color='black')
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X = pd.DataFrame(data['wt']), y = data['mpg'])
print('Intercept', model.intercept_)
print('Coef', model.coef_)
print('Độ chính xác', model.score(X = pd.DataFrame(data['wt']), y = data['mpg']))
```

```python
# pt hồi quy: y = coef*X + intercept
#data.plot(kind = 'scatter', x = 'wt', y = 'mpg', figsize = (9,9), color = 'black', xdim = (0,7))
y_pred = model.predict(X = pd.DataFrame(data['wt']))
plt.plot(data['wt'], y_pred, color = 'blue')
```

```python
# Hồi quy tuyến tính bậc 2
import numpy as np
model = linear_model.LinearRegression()
X = pd.DataFrame([data['wt'], data['wt']**2]).T
model.fit(X, y = data['mpg'])
print('Intercept', model.intercept_)
print('Coef', model.coef_)
print('Độ chính xác', model.score(X, y = data['mpg']))
range = np.arange(1.5, 5.5, 0.1)
data.plot(kind = 'scatter', x = 'wt', y = 'mpg', figsize = (9,9), color = 'black')
predictor = pd.DataFrame([range, range**2]).T
y_pred = model.predict( X = predictor)
plt.plot( range, y_pred, color = 'blue')
```

```python
# Hồi quy tuyến tính bậc 3
model = linear_model.LinearRegression()
X = pd.DataFrame([data['wt'], data['wt']**3]).T
model.fit(X, y = data['mpg'])
print('Intercept', model.intercept_)
print('Coef', model.coef_)
print('Độ chính xác', model.score(X, y = data['mpg']))
range = np.arange(1.5, 5.5, 0.1)
data.plot(kind = 'scatter', x = 'wt', y = 'mpg', figsize = (9,9), color = 'black')
predictor = pd.DataFrame([range, range**3]).T
y_pred = model.predict( X = predictor)
plt.plot( range, y_pred, color = 'blue')
```

```python
# phuong trinh hoi quy 2 bien
model = linear_model.LinearRegression()
model.fit( X=data.loc[:,['wt','hp']],y = data['mpg'])
print('intercept',model.intercept_)
print('coef',model.coef_)
print('Độ chính xác', model.score(X=data.loc[:,['wt','hp']], y=data['mpg']))
# mpg = -3.8778, wt = - 0,0317, hp = 37,227
```
