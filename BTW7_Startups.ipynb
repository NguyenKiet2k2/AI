```python
import pandas as pd
url = 'Startups.csv'
data= pd.read_csv(url)
import matplotlib.pyplot as plt
data.hist()
plt.show()
correlation= data.corr()
print(correlation)
```

```python
import seaborn as sns
sns.heatmap(correlation, annot= True)

sns.pairplot(data)
```

```python
data.plot(kind='scatter', x='R&D Spend', y='Profit',figsize=(9,9),color='black')
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X = pd.DataFrame(data['R&D Spend']), y = data['Profit'])
print('Intercept', model.intercept_)
print('Coef', model.coef_)
print('Độ chính xác', model.score(X = pd.DataFrame(data['R&D Spend']), y = data['Profit']))
```

```python
# Hồi quy tuyến tính bậc 1
model = linear_model.LinearRegression()
model.fit(X = pd.DataFrame(data['R&D Spend']), y = data['Profit'])
print('Intercept', model.intercept_)
print('Coef', model.coef_)
print('Độ chính xác', model.score(X = pd.DataFrame(data['R&D Spend']), y = data['Profit']))
data.plot(kind = 'scatter', x = 'R&D Spend', y = 'Profit', figsize = (9,9), color = 'orange')
y_pred = model.predict(X = pd.DataFrame(data['R&D Spend']))
plt.plot(data['R&D Spend'], y_pred, color = 'blue')
```

```python
# Hồi quy tuyến tính bậc 2
import numpy as np
model = linear_model.LinearRegression()
X = pd.DataFrame([data['R&D Spend'], data['Profit']**2]).T
model.fit(X, y = data['Profit'])
print('Intercept', model.intercept_)
print('Coef', model.coef_)
print('Độ chính xác', model.score(X, y = data['Profit']))
range = np.arange(0.5, 150000, 1)
data.plot(kind = 'scatter', x = 'R&D Spend', y = 'Profit', figsize = (9,9), color = 'orange')
predictor = pd.DataFrame([range, range**2]).T
y_pred = model.predict( X = predictor)
plt.plot( range, y_pred, color = 'brown')
```

```python
# Hồi quy tuyến tính bậc 3
import numpy as np
model = linear_model.LinearRegression()
X = pd.DataFrame([data['R&D Spend'], data['Profit']**3]).T
model.fit(X, y = data['Profit'])
print('Intercept', model.intercept_)
print('Coef', model.coef_)
print('Độ chính xác', model.score(X, y = data['Profit']))
range = np.arange(0.5, 150000, 1)
data.plot(kind = 'scatter', x = 'R&D Spend', y = 'Profit', figsize = (9,9), color = 'purple')
predictor = pd.DataFrame([range, range**3]).T
y_pred = model.predict( X = predictor)
plt.plot( range, y_pred, color = 'magenta')
```
